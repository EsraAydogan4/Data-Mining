```python
jb = pd.read_csv('resources/jamesbond.csv', index_col='Film')
jb.sort_index(inplace=True)

def my_sample(girilen_sayi):
  
        if girilen_sayi>0 and girilen_sayi<1:
            istVeri=int(jb.shape[0]*girilen_sayi)  #sahpe jb nin kaç satırdan ve kaç sutundan oluştuğunu gösterir
            unique_numbers = set()
            while len(unique_numbers)<istVeri:
                unique_numbers.add(rd.randrange(0,jb.shape[0]))
            myList1=list(unique_numbers)
        elif girilen_sayi>0 and girilen_sayi<jb.shape[0]:
            unique_numbers = set()
            while len(unique_numbers)<girilen_sayi:
                unique_numbers.add(rd.randrange(0,jb.shape[0]))
            myList1=list(unique_numbers)
        else: 
            return False    
        return myList1
        
jb.iloc[my_sample(5)].sort_index()
```
