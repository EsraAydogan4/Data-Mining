```python
import pandas as pd
```

```python
emp = pd.read_csv("resources/employees.csv", parse_dates=["Start Date", "Last Login Time"])
emp.info()
```

```python
emp["Last Login Time"] = pd.to_datetime(emp["Last Login Time"])
emp.info()
```

```python
emp.head
```

```python
emp["Senior Management"] = emp["Senior Management"].astype(bool)
emp.info()
```

```python
emp["Gender"].unique()
```

```python
emp["Gender"] = emp["Gender"].astype("category")
emp.info()
```

```python
emp.Team.unique()
```

```python
emp["Team"] = emp["Team"].astype("category")
emp.info() # amaç filitreleme yapmak
```

```python
emp[emp.Gender == 'Female'] # sadece famele olan kısımları getirir şartlı tarama
```

```python
emp[emp["Senior Management"]] # True yazmaya gerek yok arada boşluk varsa tıırnak kullan
```

```python
emp[emp.Salary > 70000]
```

```python
emp[emp["Start Date"] > "193.10.10"]
```

```python
emp[(emp.Team == "Finance") & (emp.Gender == "Male")]
```

```python
emp[(emp["Senior Management"]) | (emp["Start Date"] > "1993-12-01")]
```

```python
filter1 = emp["First Name"] == "Robert"
filter2 = emp.Team == "Client Services"
filter3 = emp["Start Date"] > "2016-06-01"
emp[ filter1 & filter2 | filter3]
```

```python
flt1 = emp.Team == "Legal"
flt2 = emp.Team == "Sales"
flt3 = emp.Team == "Product"

emp[flt1 | flt2 | flt3]
```

```python
emp[emp.Team.isin(["Sales", "Legal", "Product"])] #veya için sadece .isin kullanılır verileri liste biçiminde vermek gerekir
```

```python
emp[emp.Team.isnull()]
```

```python
emp[emp.Team.notnull()] #emp[-emp.Team.isnull()] aynı sonucu verir
```

```python
emp[-emp.Team.isnull()]
```

```python
emp[(emp.Salary >= 100000) & (emp.Salary <= 150000)]
```

```python
emp[emp.Salary.between(100000, 150000)] # iki sayı araındaki değerleri getirir
```

```python
emp[-emp.Salary.between(100000, 150000)] # değilini verir
```

```python
emp[emp["Start Date"].between("1993-10-01", "2000-12-30")]
```

```python
emp[emp["Last Login Time"].between("08:00 AM","06:00 PM")]
```

```python
emp[emp["First Name"].duplicated()].sort_values(by="First Name") # duplicated ikiden fazla
```

```python
emp_temp = emp.sort_values(by="First Name") 
emp_temp
```

```python
emp_temp["First Name"].duplicated() # ilk kayıt silinmez silinmeyecek değer
```

```python
emp_temp["First Name"].duplicated(keep=True) # or false or last  duplicated=tekrarlanan değerler
```

```python
emp[emp["First Name"].duplicated(keep=False)] # tüm duplice kayıtları gösterir
```

```python
emp[-emp["First Name"].duplicated(keep=False)] # dupliceolmayan kayıtları getirir örnek sınav sorusu
```
